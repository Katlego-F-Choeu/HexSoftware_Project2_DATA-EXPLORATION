```python
plt.figure(figsize = (9,9))
plt.subplots_adjust(hspace = .3)
plt.subplot(2,2,1)
plt.title('sepal length')
plt.plot(iris_data.species, iris_data.sepal_length, color = 'b', linestyle = '-', linewidth = 1)
plt.subplot(2,2,2)
plt.title('sepal width')
plt.plot(iris_data.species, iris_data.sepal_width, color = 'r', linestyle = '-', linewidth = 1)
plt.subplot(2,2,3)
plt.title('petal length')
plt.plot(iris_data.species, iris_data.petal_length, color = 'g', linestyle = '-', linewidth = 1)
plt.subplot(2,2,4)
plt.title('petal width')
plt.plot(iris_data.species, iris_data.petal_width, color = 'y', linestyle = '-', linewidth = 1)
plt.show()
```

```python
style.use('ggplot')
plt.plot(iris_data.species, iris_data.sepal_length, 'b', label = 'sepal length', linewidth = 1)
plt.plot(iris_data.species, iris_data.sepal_width, 'r', label = 'sepal width', linewidth = 1.5)
#set title of the plot
plt.title('Iris Data')
#set label for x axis 
plt.xlabel('Species')
#set label for y axis  
plt.ylabel('Sepal Length and Width')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.plot(iris_data.species, iris_data.petal_length, 'g', label = 'petal length', linewidth = 1)
plt.plot(iris_data.species, iris_data.petal_width, 'y', label = 'petal width', linewidth = 1.5)
#set title of the plot
plt.title('Iris Data')
#set label for x axis 
plt.xlabel('Species')
#set label for y axis  
plt.ylabel('Petal Length and Width')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize = (9,9))
plt.subplot(2,2,1)
plt.hist(iris_data.sepal_length, bins = 50)
plt.xlabel('Sepal')
plt.ylabel('Length')
plt.subplot(2,2,2)
plt.hist(iris_data.sepal_width, bins = 50)
plt.xlabel('Sepal')
plt.ylabel('Width')
plt.subplot(2,2,3)
plt.hist(iris_data.petal_length, bins = 50)
plt.xlabel('Petal')
plt.ylabel('Length')
plt.subplot(2,2,4)
plt.hist(iris_data.petal_width, bins = 50)
plt.xlabel('Petal')
plt.ylabel('Width')
plt.show()
```
